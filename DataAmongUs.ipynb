```python
#this sets up the figure size
plt.figure(figsize=(10, 6))

#this actually plots the graph

#Grabs the form variable, then selects the header we want to use (in this case lightswitch)
#It then counts how many times either clicky or twisty lightswitch appear
#Finally it plots a type 'bar' and give it a colour
form['lightswitch'].value_counts().plot(kind='bar', color='skyblue')

#gives it a title
plt.title('Light Switch Preference')
#give the X axis a name
plt.xlabel('Light Switch Type')
#give the Y axis a name
plt.ylabel('Count')
```

```python
form['like_sweets'].value_counts().plot(kind='bar', color='skyblue')

#gives it a title
plt.title('like_sweets')
#give the X axis a name
plt.xlabel('Do you like s')
#give the Y axis a name
plt.ylabel('Count')
```

```python
yes_sweets = form['like_sweets'].value_counts().get('Yes', 0)
no_sweets = form['like_sweets'].value_counts().get('No', 0)
print(yes_sweets, no_sweets)
```

```python
clicky_switch = form['lightswitch'].value_counts().get('Clicky light switch', 0)
twisty_switch = form['lightswitch'].value_counts().get('Twisty light switch', 0)
print(clicky_switch, twisty_switch)
```

```python
form['star_sign?'].value_counts().plot(kind='pie', color='skyblue')

#gives it a title
plt.title('Star Sign')
plt.ylabel('')
```

```python
form['Favorite_night_out'].value_counts().plot(kind='pie', color='skyblue')

#gives it a title
plt.title('Fav night out')
plt.ylabel('')
```

```python
from matplotlib_venn import venn3, venn2, venn3_circles
import scipy
import seaborn as sns
```

```python
like_sweets_yes = set(form[form['like_sweets'] == 'No'].index)
fav_night_out_at_home = set(form[form['Favorite_night_out'] == 'At home :('].index)

# Create the Venn diagram
plt.figure(figsize=(6, 6))
venn2([like_sweets_yes, fav_night_out_at_home], set_labels=('Hates Sweets', 'Favorite Night Out: At Home'))

# Add title
plt.title('Venn Diagram: Like Sweets vs Favorite Night Out At Home')

# Show the plot
plt.show()
```

```python
like_sweets_no = set(form[form['like_sweets'] == 'No'].index)
likes_clicky = set(form[form['lightswitch'] == 'Twisty Light Switch'].index)


# Create the Venn diagram
plt.figure(figsize=(6, 6))
venn2([like_sweets_no, likes_clicky], set_labels=('Hates Sweets', 'Prefers Clicky Lightswiches'))

# Add title
plt.title('Doesnt like sweets and prefers clicky light switches')

# Show the plot
plt.show()
```

```python
like_sweets_no = set(form[form['like_sweets'] == 'No'].index)
twentymiles_hike = set(form[form['20_mile_route'] == 'Hike'].index)
twentymiles_cycle = set(form[form['20_mile_route'] == 'Cycle'].index)


# Create the Venn diagram
plt.figure(figsize=(6, 6))
venn3([like_sweets_no, twentymiles_hike, twentymiles_cycle], set_labels=('Hates Sweets', 'Hikers', 'Cyclers'))

# Add title
plt.title('Doesnt like sweets and prefers clicky light switches')

# Show the plot
plt.show()
```

```python
crosstab_data = pd.crosstab(index=[form['Favorite_night_out']],
                            columns=form['like_sweets'])

plt.figure(figsize=(10, 8))

sns.heatmap(crosstab_data, annot=True, fmt="d", cmap="YlGnBu")

# Add title and labels
plt.title('Heatmap of Nights Out vs Like Sweets')
plt.xlabel('Likes Sweets')
plt.ylabel('Night Out')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
crosstab_data = pd.crosstab(index=[form['20_mile_route']],
                            columns=form['like_sweets'])

plt.figure(figsize=(10, 8))

sns.heatmap(crosstab_data, annot=True, fmt="d", cmap="YlGnBu")

# Add title and labels
plt.title('Heatmap of Prefered way to traverse 20 miles vs Like Sweets')
plt.xlabel('Likes Sweets')
plt.ylabel('20-miles ')

# Show the plot
plt.tight_layout()
plt.show()
```
